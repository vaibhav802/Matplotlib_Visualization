# Matplolib scatter plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib import style  
```

```python
data = pd.read_csv('googleplaystore.csv', nrows=1000)
```

```python
data
```

```python
x = data["Rating"]
y = data["Reviews"]
```

```python
plt.figure(dpi=100)
plt.scatter(x, y)
plt.title("Google apps scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviwes")
plt.show()
```

```python
plt.figure(dpi=100)
plt.scatter(x, y,c = "r")
plt.title("Google apps scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviwes")
plt.show()
```

```python
plt.figure(dpi=100)
plt.scatter(x, y,c = "r", marker = "+")
plt.title("Google apps scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviwes")
plt.show()
```

```python
plt.figure(dpi=100)
plt.scatter(x, y,c = "r",marker="+", s =100)
plt.title("Google apps scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviwes")
plt.show()
```

```python
plt.figure(dpi=100)
plt.scatter(x, y,c = "r",marker="+", s =100, alpha = 0.5)
plt.title("Google apps scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviwes")
plt.show()
```

```python
plt.figure(dpi=100)
plt.scatter(x, y,c = "r",marker="*", s =100, alpha = 0.5, linewidths = 10)
plt.title("Google apps scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviwes")
plt.show()
```

```python
plt.figure(dpi=100)
plt.scatter(x, y,c = "r",marker="*", s =100, alpha = 0.5, linewidths = 10, edgecolors="g")
plt.title("Google apps scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviwes")
plt.show()
```

```python
plt.figure(dpi=100)
plt.scatter(x, y,c = "r",marker="*", s =100, alpha = 0.5, linewidths = 10)
plt.title("Google apps scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviwes")
plt.show()
```

```python
plt.figure(dpi=100)
plt.scatter(x, y,c = "r",marker="*", s =100, alpha = 0.5)
plt.scatter(x, data["Installs"],c = "b",marker=".", s =100, alpha = 0.5)
plt.title("Google apps scatter plot")
plt.xlabel("Rating")
plt.ylabel("Revies & Installs")
plt.show()
```
