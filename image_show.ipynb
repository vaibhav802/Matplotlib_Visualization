# image show & color bar

```python
import matplotlib.pyplot as plt
import matplotlib.image as mi
```

```python
img = mi.imread("Chart.png")
```

```python
plt.imshow(img)
plt.show()
```

```python
plt.axis("off")
plt.imshow(img)
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.axis("off")
plt.imshow(img)
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.axis("off")
plt.imshow(img)
plt.colorbar()
plt.show()
```

```python
single = img[:,:,1]
plt.figure(figsize=(15,6))
plt.axis("off")
plt.imshow(single, cmap = "hot")
plt.colorbar()
plt.show()
```
