# Matplotlib subplots


```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(2,2,1)
plt.pie([1])
plt.subplot(2,2,2)
plt.pie([1,2])
plt.subplot(2,2,3)
plt.pie([1,2,3])
plt.subplot(2,2,4)
plt.pie([1,2,3,4])
plt.show()
```

```python
plt.figure(figsize=(12,8))

#plt.subplot(3,2,1)
plt.subplot(321)
plt.plot([1,2,3])
plt.title("line plot")

plt.subplot(3,2,2)
plt.hist([1,2,44])
plt.title("Histogram")

plt.subplot(3,2,3)
plt.bar([1,4,5],[3,6,2])
plt.title("Bar Chart")

plt.subplot(3,2,4)
plt.pie([1,2,4])
plt.title("pie Chart")

plt.subplot(3,2,5)
plt.scatter([1,24,3], [4,32,1])
plt.title("scatter Plot")

plt.subplot(3,2,6)
plt.barh([2,2,3], [2,4,5])
plt.title("Horizontal bar chart")
plt.show()
```
