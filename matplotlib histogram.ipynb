## Matplotlib Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
import random
from matplotlib import style
```

```python
ml_age = np.random.randint(18,45, (100))
py_age = np.random.randint(20,50, (100))
```

```python
ml_age
```

```python
py_age
```

### simple hist plot

```python
plt.hist(ml_age)
plt.title("Ml Student Age Histogram")
plt.xlabel("Student Age")
plt.ylabel("No.of Student Age")
plt.show()
```

###  Use of multiple parameter

```python
plt.hist(ml_age, bins = [15,20,25,30,35,40,45], rwidth=0.8)
plt.title("Ml Student Age Histogram")
plt.xlabel("Student Age")
plt.ylabel("No.of Student Age")
plt.show()
```

```python
plt.hist(ml_age, bins = [15,20,25,30,35,40,45], rwidth=0.8, histtype=  "step")
plt.title("Ml Student Age Histogram")
plt.xlabel("Student Age")
plt.ylabel("No.of Student Age")
plt.show()
```

```python
plt.hist(ml_age, bins = [15,20,25,30,35,40,45], rwidth=0.8, align = 'right')
plt.title("Ml Student Age Histogram")
plt.xlabel("Student Age")
plt.ylabel("No.of Student Age")
plt.show()
```

```python
plt.hist(ml_age, bins = [15,20,25,30,35,40,45], rwidth=0.8, orientation="horizontal")
plt.title("Ml Student Age Histogram")
plt.xlabel("Student Age")
plt.ylabel("No.of Student Age")
plt.show()
```

```python
plt.hist(ml_age, bins = [15,20,25,30,35,40,45], rwidth=0.8, color = "r")
plt.title("Ml Student Age Histogram")
plt.xlabel("Student Age")
plt.ylabel("No.of Student Age")
plt.show()
```

```python
plt.hist(ml_age, bins = [15,20,25,30,35,40,45], rwidth=0.8, label = "Ml age")
plt.title("Ml Student Age Histogram")
plt.xlabel("Student Age")
plt.ylabel("No.of Student Age")
plt.legend()
plt.show()
```

```python
plt.hist(ml_age, bins = [15,20,25,30,35,40,45], rwidth=0.8 , stacked="False")
plt.title("Ml Student Age Histogram")
plt.xlabel("Student Age")
plt.ylabel("No.of Student Age")
plt.show()
```

```python
plt.hist(ml_age, bins = [15,20,25,30,35,40,45], rwidth=0.8, color = "r", label = "Ml_student")
plt.hist(py_age, bins = [15,20,25,30,35,40,45], rwidth=0.8, color = "b", label = "Py_student")
plt.title("Ml Student Age Histogram")
plt.xlabel("Student Age")
plt.ylabel("No.of Student Age")
plt.legend()
plt.show()
```

```python
plt.hist(ml_age, bins = [15,20,25,30,35,40,45], rwidth=0.8, color = "r", label = "Ml_student", orientation="vertical")
plt.hist(py_age, bins = [15,20,25,30,35,40,45], rwidth=0.8, color = "b", label = "Py_student", orientation= "horizontal")
plt.title("Ml Student Age Histogram")
plt.xlabel("Student Age")
plt.ylabel("No.of Student Age")
plt.legend()
plt.show()
```

### Change plot size and background

```python
plt.figure(figsize = (14,8))
plt.hist([ml_age, py_age], bins = [15,20,25,30,35,40,45], rwidth=0.8, color = ["r","b"], label = ["Ml_student","Py_student"]
        ,orientation="vertical", align = "right")
#plt.hist(py_age, bins = [15,20,25,30,35,40,45], rwidth=0.8, color = "b", label = "Py_student", orientation= "vertical")
plt.title("Ml Student Age Histogram")
plt.xlabel("Student Age")
plt.ylabel("No.of Student Age")
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (14,8))
style.use("ggplot")
plt.hist([ml_age, py_age], bins = [15,20,25,30,35,40,45], rwidth=0.8, color = ["r","b"], label = ["Ml_student","Py_student"]
        ,orientation="vertical", align = "right")
#plt.hist(py_age, bins = [15,20,25,30,35,40,45], rwidth=0.8, color = "b", label = "Py_student", orientation= "vertical")
plt.title("Ml Student Age Histogram")
plt.xlabel("Student Age")
plt.ylabel("No.of Student Age")
plt.legend()
plt.show()
```
