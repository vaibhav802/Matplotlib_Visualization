# Matplotlib_pie_chart


```python
import matplotlib.pyplot as plt
```

```python
classes = ['Python', 'R', 'Ml', 'AI', 'Ruby', 'Pearl', 'Data Science']
class1  = [34, 56, 33 ,52, 45 ,43, 50]
```

```python
plt.figure(dpi=100)
plt.pie(class1)
plt.show()
```

```python
plt.figure(dpi=100)
plt.pie(class1, labels = classes)
plt.show()
```

```python
plt.figure(dpi=100)
plt.pie(class1, labels = classes, explode = [0,0.1,0.1,0.1,0.1,0,0])
plt.show()
```

```python
plt.figure(dpi=100)
plt.pie(class1, labels = classes, colors = ['c','b','g','r','b','y','r'] )
plt.title('scatter plot')
plt.show()
```

```python
plt.figure(dpi=100)
plt.pie(class1, labels = classes, autopct="%0.1f%%")
plt.show()
```

```python
plt.figure(dpi=100)
plt.pie(class1, labels = classes, shadow=True)
plt.show()
```

```python
plt.figure(dpi=100)
plt.pie(class1, labels = classes, radius = 1)
plt.show()
```

```python
plt.figure(dpi=100)
plt.pie(class1, labels = classes, autopct="%0.1f%%", startangle=50, textprops={"fontsize":10})
plt.show()
```

```python
plt.figure(dpi=100)
plt.pie(class1, labels = classes, autopct="%0.1f%%", pctdistance=0.5)
plt.show()
```

```python
plt.figure(dpi=100)
plt.pie(class1, labels = classes, autopct="%0.1f%%", labeldistance=1.3)
plt.show()
```

```python
plt.figure(dpi=100)
plt.pie(class1, labels = classes, autopct="%0.1f%%",wedgeprops={'linewidth': 2, "width":1, "edgecolor":"r"})
plt.show()
```

```python
plt.figure(dpi=100)
plt.pie(class1, labels = classes, autopct="%0.1f%%", center=(2,3))
plt.show()
```

```python
plt.figure(dpi=100)
plt.pie(class1, labels = classes, autopct="%0.1f%%",center = (5,3))
plt.legend(loc =9)
plt.show()
```
