## Matplotlib Bar Chart


```python
import matplotlib.pyplot as plt
from matplotlib import style 
import numpy as np
```

```python
classes = ['Python', 'R', 'AI', 'ML', 'DS']
class1 = [30, 10, 20, 25, 10]
class2 = [40, 5 ,20, 20, 10]
class3 = [35, 5, 30, 15, 15]
```

```python
plt.bar(classes , class1)
```

```python
plt.barh(classes , class2)
```

```python
plt.bar(classes , class1 , width=0.3)
```

```python
plt.bar(classes , class1 , align = "edge")
```

```python
plt.bar(classes , class1 , color = "r",
       edgecolor= "b", linewidth = 5)
```

```python
plt.bar(classes , class1 , color = "r",
       edgecolor= "b", linewidth = 5, alpha = 0.5)
```

```python
plt.bar(classes , class1 , color = "br",
       edgecolor= "k", linewidth = 5, linestyle = "--")
```

```python
plt.bar(classes , class1 , color = "br",
       edgecolor= "k", linewidth = 5, label = "class 1")
plt.legend()
plt.show()
```

```python
plt.bar(classes , class1 , color = "yr",
       edgecolor= "m", linewidth = 5, label = "class 1", visible = False)
plt.legend()
plt.show()
```

###  Multiple Bar chart 

```python
style.use("ggplot")
plt.figure(dpi = 100)
plt.figure(figsize= (10,5))
plt.bar(classes , class1 , color = "r",
       edgecolor= "k", linewidth = 5, label = "class 1")
plt.legend()
plt.show()
```

```python
style.use("ggplot")
plt.figure(dpi = 100)
plt.figure(figsize= (10,5))
plt.barh(classes , class1 , color = "r",
       edgecolor= "k", linewidth = 5, label = "class 1")
plt.legend()
plt.show()
```

```python
style.use("ggplot")
plt.figure(dpi = 100)
plt.figure(figsize= (10,5))
plt.bar(classes , class1 , color = "r",
     linewidth = 5, label = "class 1")
plt.title("classes", fontsize = 14)
plt.xlabel("class1")
plt.ylabel("No of student")
plt.legend()
plt.show()
```

```python
style.use("ggplot")
plt.figure(dpi = 100)
plt.figure(figsize= (10,5))
plt.bar(classes , class1 , color = "r",linewidth = 5, label = "class 1")
plt.bar(classes , class2 , color = "b",linewidth = 5, label = "class 2")
plt.title("classes", fontsize = 14)
plt.xlabel("classes")
plt.ylabel("No of student")
plt.legend()
plt.show()
```

```python
style.use("ggplot")
plt.figure(dpi = 100)
plt.figure(figsize= (10,5))
classes_index = np.arange(len(classes))
width = 0.2
plt.bar(classes_index , class1 , width , color = "r", label = "class 1")
plt.bar(classes_index + width, class2 ,width, color = "b", label = "class 2")
plt.title("classes", fontsize = 14)
plt.xlabel("classes")
plt.ylabel("No of student")
plt.legend()
plt.show()
```

```python
style.use("ggplot")
plt.figure(dpi = 100)
plt.figure(figsize= (10,5))
classes_index = np.arange(len(classes))
width = 0.2
plt.bar(classes_index , class1 , width , color = "r", label = "class 1")
plt.bar(classes_index + width, class2 ,width, color = "b", label = "class 2")
plt.title("classes", fontsize = 14)
plt.xticks(classes_index, classes)
plt.xlabel("classes")
plt.ylabel("No of student")
plt.legend()
plt.show()
```

```python
style.use("ggplot")
plt.figure(dpi = 100)
plt.figure(figsize= (10,5))
classes_index = np.arange(len(classes))
width = 0.2
plt.bar(classes_index , class1 , width , color = "r", label = "class 1")
plt.bar(classes_index + width, class2 ,width, color = "b", label = "class 2")
plt.title("classes", fontsize = 14)
plt.xticks(classes_index, classes, rotation = 20)
plt.xlabel("classes")
plt.ylabel("No of student")
plt.legend()
plt.show()
```

```python
style.use("ggplot")
plt.figure(dpi = 100)
plt.figure(figsize= (10,5))
classes_index = np.arange(len(classes))
width = 0.2
plt.bar(classes_index , class1 , width , color = "r", label = "class 1")
plt.bar(classes_index + width, class2 ,width, color = "b", label = "class 2")
plt.bar(classes_index + width+width, class3 ,width, color = "c", label = "class 3")
plt.title("classes", fontsize = 14)
plt.xticks(classes_index, classes, rotation = 20)
plt.xlabel("classes")
plt.ylabel("No of student")
plt.legend()
plt.show()
```

```python
style.use("ggplot")
plt.figure(dpi = 100)
plt.figure(figsize= (10,5))
classes_index = np.arange(len(classes))
width = 0.2
plt.barh(classes_index , class1 , width , color = "r", label = "class 1")
plt.barh(classes_index + width, class2 ,width, color = "b", label = "class 2")
plt.barh(classes_index + width+width, class3 ,width, color = "c", label = "class 3")
plt.title("classes", fontsize = 14)
plt.yticks(classes_index, classes, rotation = 20)
plt.ylabel("classes")
plt.xlabel("No of student")
plt.legend()
plt.show()
```
