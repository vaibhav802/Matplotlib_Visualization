## Line Plot


```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
days = np.arange(1,16)
temperature = np.arange(35.6,65,2)
mum_tem = np.arange(37.2 ,66,2)
```

### simple line plot

```python
plt.plot(days, temperature)
plt.title('Days vs temperature')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.axis([0, 30, 30, 70])
plt.show()
```

###  Change line and marker parameter

```python
plt.plot(days, temperature, color = 'r', marker = "o", linestyle = "--", linewidth = 2, 
         markersize = 5)
plt.title('Days vs temperature')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.axis([0, 30, 30, 70])
plt.show()
```

### Add labels

```python
plt.plot(days, temperature, "ro--", linewidth = 2, 
         markersize = 5, label = 'Temperature Line')
plt.title('Days vs temperature', fontsize = 15)
plt.xlabel('Days', fontsize = 15)
plt.ylabel('Temperature', fontsize = 15)
plt.axis([0, 30, 30, 70])
plt.legend(["Temperature Line"], loc = 2 )
plt.show()
```

### Add griding

```python
style.use("ggplot")
plt.plot(days, temperature, "mo--", linewidth = 2, 
         markersize = 5, label = 'Temperature Line')
plt.title('Days vs temperature', fontsize = 15)
plt.xlabel('Days', fontsize = 15)
plt.ylabel('Temperature', fontsize = 15)
plt.axis([0, 30, 30, 70])
plt.legend()
plt.grid(color='k', linestyle='-', linewidth=1)
plt.show()
```

### mupliline graph 

```python
style.use("ggplot")
plt.plot(days, temperature, "mo--", linewidth = 2, 
         markersize = 5, label = 'Temperature Line')
plt.plot(days, mum_tem, "yo--", linewidth = 2, 
         markersize = 5, label = 'Mumbai Line')
plt.title('Temperature', fontsize = 15)
plt.xlabel('Days', fontsize = 15)
plt.ylabel('Temperature', fontsize = 15)
plt.axis([0, 30, 30, 70])
plt.grid(color='k', linestyle='-', linewidth=1)
plt.legend()
plt.show()
```
